## Coordinate Time

```julia
include("compute_BL_analytic_inspiral.jl")
```

### Plot constants of motion

```julia
# load trajectory
t_BL_Analytic, r_BL_Analytic, θ_BL_Analytic, ϕ_BL_Analytic, dr_dt_BL_Analytic, dθ_dt_BL_Analytic,
dϕ_dt_BL_Analytic, d2r_dt2_BL_Analytic, d2θ_dt2_BL_Analytic, d2ϕ_dt2_BL_Analytic, dt_dτ_BL_Analytic = 
InspiralEvolution.Analytic.BLTime.load_trajectory(a, p, e, θmin, q, kerrReltol, data_path);

t_BL_Analytic = t_BL_Analytic * MtoSecs;

t_Fluxes_BL_Analytic, EE_BL_Analytic, Edot_BL_Analytic, LL_BL_Analytic, Ldot_BL_Analytic, QQ_BL_Analytic, Qdot_BL_Analytic, CC_BL_Analytic, Cdot_BL_Analytic,
pArray_BL_Analytic, ecc_BL_Analytic, θmin_BL_Analytic = 
InspiralEvolution.Analytic.BLTime.load_constants_of_motion(a, p, e, θmin, q, kerrReltol,
data_path);
t_Fluxes_BL_Analytic = t_Fluxes_BL_Analytic * MtoSecs;

ι_BL_Analytic = @. acos(LL_BL_Analytic / sqrt(LL_BL_Analytic^2 + CC_BL_Analytic));
```

```julia
# some plot attributes
BL_Analytic_col = :red;
BL_Analytic_ls = :solid;
BL_Analytic_lw=1.0;
# some plot attributes
BL_col = :red
BL_ls = :solid
BL_lw = 2.0
wave_im_ratio = 5   # width / height
wave_height = 300; wave_width = wave_height * wave_im_ratio
xtickfontsize=10; ytickfontsize=10; guidefontsize=15;
# margins
left_margin = 8mm; right_margin=8mm; top_margin=8mm; bottom_margin=8mm;
```

```julia
gr()
plotE = Plots.plot(t_Fluxes_BL_Analytic, 1e4 * (EE_BL_Analytic .- EE_BL_Analytic[1]), ylabel=L"E", legend=:false, 
    color=BL_Analytic_col, linestyle=BL_Analytic_ls,
    left_margin =  left_margin,
    right_margin =  right_margin,
    top_margin =  top_margin,
    bottom_margin =  bottom_margin)

plotL = Plots.plot(t_Fluxes_BL_Analytic, LL_BL_Analytic, ylabel=L"L", legend=:false,
    color=BL_Analytic_col, linestyle=BL_Analytic_ls,
    left_margin =  left_margin,
    right_margin =  right_margin,
    top_margin =  top_margin,
    bottom_margin =  bottom_margin)


plotC = Plots.plot(t_Fluxes_BL_Analytic, CC_BL_Analytic, ylabel=L"C", legend=:false,
    color=BL_Analytic_col, linestyle=BL_Analytic_ls,
    left_margin =  left_margin,
    right_margin =  right_margin,
    top_margin =  top_margin,
    bottom_margin =  bottom_margin)

plotP = Plots.plot(t_Fluxes_BL_Analytic, pArray_BL_Analytic, ylabel=L"p", legend=:false, 
    color=BL_Analytic_col, linestyle=BL_Analytic_ls,
    left_margin =  left_margin,
    right_margin =  right_margin,
    top_margin =  top_margin,
    bottom_margin =  bottom_margin)

plotEcc = Plots.plot(t_Fluxes_BL_Analytic, ecc_BL_Analytic, ylabel=L"e", xlabel=L"t\ (s)", legend=:false,
    color=BL_Analytic_col, linestyle=BL_Analytic_ls,
    left_margin =  left_margin,
    right_margin =  right_margin,
    top_margin =  top_margin,
    bottom_margin =  bottom_margin)

plotι = Plots.plot(t_Fluxes_BL_Analytic, 1e4 * (ι_BL_Analytic .- ι_BL_Analytic[1]), ylabel=L"\iota", legend=:false,
    color=BL_Analytic_col, linestyle=BL_Analytic_ls,
    left_margin =  left_margin,
    right_margin =  right_margin,
    top_margin =  top_margin,
    bottom_margin =  bottom_margin)

orbitalParamsPlot=plot(plotE, plotL, plotC, plotP, plotEcc, plotι, layout = @layout([[A B C]; [D E F]]),
size=(1500, 600), dpi=100)
display("image/png", orbitalParamsPlot)
```
